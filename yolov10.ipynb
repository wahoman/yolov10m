```python
# 필요한 라이브러리 임포트
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# 모델 로드 (best.pt 경로를 정확하게 지정하세요)
model = YOLO('runs/detect/train/weights/best.pt')

# 예측하려는 이미지 경로 (이미지 경로를 정확하게 지정하세요)
image_path = '/content/{HOME}/datasets/x-ray-1/train/images/augmented_12_jpg.rf.09dacdeac9b57bdd5df9636695277396.jpg'

# 예측 수행
results = model.predict(source=image_path, save=True)  # save=True를 통해 시각화 결과 저장

# 결과 확인
print(f"Results: {results}")

# 첫 번째 결과에서 시각화된 이미지 가져오기
result_image_path = results[0].save_path  # 저장된 이미지 경로
result_image = cv2.imread(result_image_path)

# 결과 이미지 표시
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
